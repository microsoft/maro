## Facility info: name, storage capacity

```python
df_facility
```

```python
def facility_name2ref(name: str):
    if name.startswith("STRG"):
        return "WarehouseFacility"
    elif name.startswith("PLANT"):
        return "PlantFacility"
    else:
        return "StoreFacility"
```

```python
# Key: facility name
facility_info: Dict[str, dict] = {}
```

```python
for facility_name, capacity in zip(
    df_facility["WarehouseId"].to_numpy(),
    df_facility["StorageCapacity"].to_numpy(),
):
    ref = facility_name2ref(facility_name)
    facility_info[facility_name] = {
        "ref": ref,
        "storage_capacity": int(capacity),
        "skus": {},
    }
```

```python
facility_info
```

## Convert facility info to string

```python
def get_vendor_string():
    config = (
        f"    - name: Super_Vendor\n"
        f"      definition_ref: SuperVendor\n"
        f"      skus:\n"
    )
    
    vendor_skus = df_item[~df_item["VendorId"].isnull()]["ItemId"].unique()
    for sku_name in vendor_skus:
        config += (
            f"        {sku_name}:\n"
            f"          price: 0.0\n"
            f"          init_stock: 10000\n"  # TODO: replace with another value, or set default value for it.
            
        )
    
    config += (
        f"      children:\n"
        f"        storage:\n"
        f"          config:\n"
        f"            capacity: null\n"
        f"            unit_storage_cost: 0\n"
        f"        distribution:\n"
        f"          config:\n"
        f"            ROUTE_V:\n"
        f"              number: null\n"
        f"      config:\n"
        f"        unit_order_cost: 0\n"
        f"        unit_delay_order_penalty: 0\n"
        f"\n"
    )

    return config
```

```python
def convert_facility_info_to_string(name, info):
    config = (
        f"    - name: {name}\n"
        f"      definition_ref: {info['ref']}\n"
        f"      skus:\n"
    )

    for sku_name, sku_info in info["skus"].items():
        config += (
            f"        {sku_name}:\n"
            f"          price: {sku_info['price']}\n"
            f"          init_stock: {sku_info['init_stock']}\n"
            f"          unit_storage_cost: {sku_info['unit_storage_cost']}\n"
            f"          has_consumer: True\n"
        )
        
        if "manufacture" in sku_info:
            config += (
                f"          has_manufacture: True\n"
                f"          unit_product_cost: {sku_info['manufacture']['unit_cost']}\n"
                f"          max_manufacture_rate: {sku_info['manufacture']['max_rate']}\n"
                f"          manufacture_leading_time: {sku_info['manufacture']['leading_time']}\n"
            )
        
        if "seller" in sku_info:
            config += (
                f"          has_seller: True\n"
                f"          sale_gamma: {sku_info['seller']['gamma']}\n"
                f"          backlog_ratio: 0.0\n"
            )

    config += (
        f"      children:\n"
        f"        storage:\n"
        f"          config:\n"
        f"            capacity: {info['storage_capacity']}\n"
    )
    
    if "distribution" in info:
        config += (
            f"        distribution:\n"
            f"          config:\n"
        )
        for vehicle in info["distribution"]:
            config += f"            {vehicle[0]}:\n"
            if np.isnan(vehicle[1]):
                config += f"              number: null\n"  # null indicate infinite distribution capacity
            else:
                config += f"              number: {int(vehicle[1])}\n"

    config += (
        f"      config:\n"
        f"        unit_order_cost: {info['unit_order_cost']}\n"
        f"        unit_delay_order_penalty: 0\n"  # Notes: use 0 delay order penalty
    )
    if info['ref'] == "StoreFacility":
        config += (
            f"        file_path: {info['sales_path']}\n"
        )
    
    config += "\n"
    
    return config
```

```python
config_facilities = get_vendor_string()

for facility_name, info in facility_info.items():
    if info['ref'] == "PlantFacility":
        config_facilities += convert_facility_info_to_string(facility_name, info)

for facility_name, info in facility_info.items():
    if info['ref'] == "WarehouseFacility":
        config_facilities += convert_facility_info_to_string(facility_name, info)

for facility_name, info in facility_info.items():
    if info['ref'] == "StoreFacility":
        config_facilities += convert_facility_info_to_string(facility_name, info)
```
